```python
#Image training and test data generator
gen1 = ImageDataGenerator(rotation_range=5,rescale=1.0/255,horizontal_flip=True)
gen2 = ImageDataGenerator(rescale=1.0/255)

train_generator = gen1.flow_from_directory('data/labelled_data/train/',target_size=(50,50),batch_size=50)
test_generator = gen1.flow_from_directory('data/labelled_data/test/',target_size=(50,50),batch_size=10)
validation_generator = gen1.flow_from_directory('data/labelled_data/train/',target_size=(50,50),batch_size=10)
```

```python
#Models
from keras.layers import Conv2D,MaxPool2D,Flatten,Dense, Dropout
from keras.models import Sequential
```

```python
NUM_STUDENTS = 7
```

```python
def get_model1():
    model = Sequential()
    #Convolutional Network
    model.add(Conv2D(32,(3,3),activation='relu',input_shape=(50,50,3)))
    model.add(Conv2D(64,(3,3),activation='relu'))
    model.add(MaxPool2D(pool_size=(2,2)))
    model.add(Conv2D(32,(3,3),activation='relu'))
    model.add(Conv2D(64,(3,3),activation='relu'))
    model.add(Dropout(0.3))
    model.add(MaxPool2D(pool_size=(2,2)))
    model.add(Conv2D(32,(3,3),activation='relu'))
    model.add(Conv2D(64,(3,3),activation='relu'))
    model.add(Dropout(0.2))
    model.add(MaxPool2D(pool_size=(2,2)))

    #Dense network
    model.add(Flatten())
    model.add(Dense(512,activation='relu'))
    model.add(Dense(1024,activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(1024,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(NUM_STUDENTS,activation='softmax'))
    return model
```

```python
def get_model2():
    #VGG-16 like model
    model = Sequential()
    model.add(Conv2D(32,(3,3),activation='relu',input_shape=(50,50,3)))
    model.add(Conv2D(32,(3,3),activation='relu'))
    model.add(MaxPool2D(pool_size=(2,2)))
    
    model.add(Conv2D(64,(3,3),activation='relu'))
    model.add(Conv2D(64,(3,3),activation='relu'))
    model.add(MaxPool2D(pool_size=(2,2)))
    
    model.add(Conv2D(128,(3,3),activation='relu'))
    model.add(Conv2D(128,(3,3),activation='relu'))
    model.add(MaxPool2D(pool_size=(2,2)))
    
    model.add(Flatten())
    model.add(Dense(1024,activation='relu'))
    model.add(Dense(1024,activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(1024,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(NUM_STUDENTS,activation='softmax'))
    return model
```

```python
from keras.optimizers import Adam
```

```python
model = get_model2()
optimizer = Adam(lr = 0.0001)
model.compile(optimizer,loss="categorical_crossentropy",metrics=['acc'])
```

```python
#Har Like Face detector
HAR_CLASSIFIER = cv2.CascadeClassifier('data/haar_train_data.xml')
```

```python
def convert_to_class_name(predictions,indices):
    class_name_preds = []
    for p in predictions:
        for k in indices.keys():
            if indices[k] == p:
                class_name_preds.append(k)
    return class_name_preds
```

```python
from os import listdir
def load_images(path):
    files = listdir(path)
    images = []
    for f in files:
        img = converToRGB(cv2.imread(path+'/'+f))
        images.append(img)
    return images
```

```python
def plot_images(images):
    fig, axes = plt.subplots(1,len(images))
    for i in range(len(images)):
        if len(images) == 1:
            axes.imshow(images[i])
            axes.get_xaxis().set_visible(False)
            axes.get_yaxis().set_visible(False)
            break
        axes[i].imshow(images[i])
        axes[i].get_xaxis().set_visible(False)
        axes[i].get_yaxis().set_visible(False)
```

```python
def detect_faces_har(image):
    """
    image: colored image
    Return: returns bounding boxes for faces
    """
    gray_image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    faces = HAR_CLASSIFIER.detectMultiScale(gray_image,scaleFactor = 1.2, minNeighbors=5)
    return faces
```

```python
def crop_to_faces(images):
    cropped_faces = []
    for i in images:
        locations = detect_faces_har(i)
        for (x,y,w,h) in locations:
            cropped_image = i[y:y+h, x:x+w]
            cropped_faces.append(cv2.resize(cropped_image,(50,50)))
    return cropped_faces
```
